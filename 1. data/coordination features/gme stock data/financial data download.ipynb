```python
"""
Simple Alpaca Data Downloader
==============================
Downloads stock data and saves to CSV.

Requirements:
    pip install alpaca-trade-api pandas
"""

import pandas as pd
from alpaca_trade_api.rest import REST, TimeFrame

# ============================================================
# 在这里设置您的API密钥 (Set your API keys here)
# ============================================================
API_KEY = "PKNEE7BCSX466DHEWZVI62L47R"
SECRET_KEY = "Ct7QpRCqXobspnaxRTb3koRkrKS6XatSWFwwY2rMGu3m"
# ============================================================


def download_stock_data(
    symbol='GME',
    start_date="2019-07-01",
    end_date="2021-06-30",
    timeframe='5Min',
    output_file='stock_data.csv'
):
    """
    Download stock data from Alpaca and save to CSV.
    
    Parameters:
    -----------
    symbol : str
        Stock symbol (e.g., 'GME', 'AAPL', 'TSLA')
    start_date : str
        Start date in 'YYYY-MM-DD' format
    end_date : str
        End date in 'YYYY-MM-DD' format
    timeframe : str
        Time interval - Options:
        '1Min'  - 1 minute bars
        '5Min'  - 5 minute bars
        '15Min' - 15 minute bars
        '30Min' - 30 minute bars
        '1Hour' - 1 hour bars
        '1Day'  - Daily bars
    output_file : str
        Output CSV filename
    """
    
    print(f"\nDownloading {symbol} data...")
    print(f"Period: {start_date} to {end_date}")
    print(f"Timeframe: {timeframe}")
    
    # Connect to Alpaca API
    api = REST(
        key_id=API_KEY,
        secret_key=SECRET_KEY,
        base_url='https://paper-api.alpaca.markets'
    )
    
    # Map timeframe - using string format for compatibility
    timeframe_map = {
        '1Min': '1Min',
        '5Min': '5Min',
        '15Min': '15Min',
        '30Min': '30Min',
        '1Hour': '1Hour',
        '1Day': '1Day'
    }
    
    tf = timeframe_map[timeframe]
    
    # Download data
    bars = api.get_bars(
        symbol,
        tf,
        start=start_date,
        end=end_date,
        adjustment='raw'
    ).df
    
    # Reset index to make timestamp a column
    bars = bars.reset_index()
    
    # Save to CSV
    bars.to_csv(output_file, index=False)
    
    print(f"\n✓ Success!")
    print(f"  Downloaded {len(bars):,} bars")
    print(f"  Saved to: {output_file}")
    print(f"  Columns: {', '.join(bars.columns)}")
    print(f"  Date range: {bars['timestamp'].min()} to {bars['timestamp'].max()}")
    print(f"  Price range: ${bars['low'].min():.2f} - ${bars['high'].max():.2f}")
    
    return bars


if __name__ == "__main__":
    
    # Check API keys
    if API_KEY == 'your_api_key_here':
        print("\n" + "="*60)
        print("ERROR: Please set your API keys first!")
        print("="*60)
        print("\nSteps:")
        print("1. Sign up at https://alpaca.markets")
        print("2. Get your API keys from dashboard")
        print("3. Edit this file and replace:")
        print("   API_KEY = 'your_api_key_here'")
        print("   SECRET_KEY = 'your_secret_key_here'")
        print("="*60)
    else:
        # Download GameStop data - 5 minute bars
        download_stock_data(
            symbol='GME',
            start_date="2019-07-01",
            end_date="2021-06-30",
            timeframe='5Min',
            output_file='gme_5min.csv'
        )
        
        # Download S&P 500 ETF (SPY) - 5 minute bars
        download_stock_data(
            symbol='SPY',
            start_date="2019-07-01",
            end_date="2021-06-30",
            timeframe='5Min',
            output_file='spy_5min.csv'
        )
        
        # Download Dow Jones ETF (DIA) - 5 minute bars
        download_stock_data(
            symbol='DIA',
            start_date="2019-07-01",
            end_date="2021-06-30",
            timeframe='5Min',
            output_file='dia_5min.csv'
        )
        
        print("\n" + "="*60)
        print("✓ All downloads complete!")
        print("="*60)
        print("\nFiles created:")
        print("  - gme_5min.csv  (GameStop)")
        print("  - spy_5min.csv  (S&P 500 ETF)")
        print("  - dia_5min.csv  (Dow Jones ETF)")
        print("="*60)
        
       
```
